# Manual Statistics in Python

This notebook demonstrates how to compute the **mean**, **standard deviation**, **coefficient of variation**, and **percentiles** manually without using NumPy or Pandas.

```python
# Sample data
data = [10, 20, 30, 40, 50]
```

```python
# Mean
def calculate_mean(data):
    return sum(data) / len(data)

mean = calculate_mean(data)
mean
```

```python
# Standard Deviation (Sample)
def calculate_std(data):
    mean = calculate_mean(data)
    squared_diffs = [(x - mean) ** 2 for x in data]
    return (sum(squared_diffs) / (len(data) - 1)) ** 0.5

std = calculate_std(data)
std
```

```python
# Coefficient of Variation
def coefficient_of_variation(data):
    return calculate_std(data) / calculate_mean(data)

cv = coefficient_of_variation(data)
cv
```

```python
# Percentile
def percentile(data, percent):
    data_sorted = sorted(data)
    k = (len(data_sorted)-1) * (percent / 100)
    f = int(k)
    c = k - f
    if f + 1 < len(data_sorted):
        return data_sorted[f] + (data_sorted[f+1] - data_sorted[f]) * c
    else:
        return data_sorted[f]

# Example: 25th, 50th, 75th percentile
p25 = percentile(data, 25)
p50 = percentile(data, 50)
p75 = percentile(data, 75)

p25, p50, p75
```
