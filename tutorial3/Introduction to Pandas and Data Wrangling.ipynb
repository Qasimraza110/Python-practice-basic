# Introduction to Pandas and Data Wrangling

## What is Pandas?
Pandas is a powerful open-source Python library used for data analysis and manipulation. It provides two main data structures:

* Series: One-dimensional labeled array.

* DataFrame: Two-dimensional labeled table (like an Excel sheet).

Pandas makes it easier to load, manipulate, clean, and analyze data. It is especially useful when working with CSV, Excel files, SQL databases, or even JSON.

## What is Data Wrangling?
Data wrangling, also known as data munging, is the process of cleaning and transforming raw data into a usable format. This includes:

* Removing or correcting invalid data

* Filling in missing values

* Transforming data types

* Merging and reshaping datasets

This line tells us how many missing values are present in each column of the `amazonbooks` DataFrame.

* 🧮 Now, it sums up the True values column-wise:

* In Python and pandas:

   * `True` is automatically treated as 1

   * `False` is automatically treated as 0

* So `.sum()` just adds the 1s in each column → which gives us the count of missing values per column.

✅ Boolean coercion:
Python treats `True` as `1` and `False` as `0` in mathematical operations.

✅ `.sum()` on a DataFrame:
By default, it adds values down each column (`axis=0`), which is what we want here.

If you use `axis=1`, it would add across each row (which is useful in other cases).

## 🧠 What is .iloc?
* `.iloc[]` stands for “integer-location based indexing”

* It lets you select rows and columns by their numerical position, not by name.

Think of it like saying: “Give me the first few rows and columns by number, not by label.”

```python
amazonbooks.iloc[:5,:] # amazonbooks.iloc[0:5,0:13]
```

This means:

* `:5` → rows from index 0 up to (but not including) 5, i.e., the first 5 rows

* `:` → all columns

✅ Result: The first 5 rows and all columns of the amazonbooks DataFrame.

💡 Pro Tip:
* Use `.iloc[::]` to slice with row and column numbers

* Use `.loc[::]` to slice with row and column labels

### How to detect rows with multiple missing values using a logical condition.

1. amazonbooks.isnull()
   * Returns a DataFrame of the same shape as amazonbooks

   * Each cell becomes:

    * `True` if the original value is missing (NaN)

    * `False` if the value exists

2. `.sum(axis=1)`
   * This sums the True values row-wise (`axis=1` means across columns)

   * Since `True = 1` and `False = 0`, this gives the number of missing values in each row

3. `> 1`
   * Compares each row’s missing value count

   * Returns:

    * `True` for rows with more than one missing value

    * `False` otherwise

✅ Final Result: A Boolean Series where:

* Each index corresponds to a row in amazonbooks

* The value is True if that row has more than one missing value

```python
amazonbooks.isnull().sum(axis=1)>1
```

🧠 Why is this useful?
In data wrangling, we often want to:

* Flag incomplete records

* Decide whether to drop or impute them

This line is a quick filter for finding rows that might be too incomplete to use directly.



---



🧠 Understanding the logic inside .iloc[ ... , :]

`amazonbooks.isnull().sum(axis=1) > 1`

* As explained earlier, this returns a Boolean Series

* Each `True` indicates a row with more than one missing value

🧠 Passing the Boolean Series to .iloc

`amazonbooks.iloc[ boolean_series , : ]`

* `.iloc[]` is designed for integer positions only, not labels or Boolean masks.

* So this line raises an error (that’s why it says # breaks!)

* It tries to use the Boolean Series with .iloc, but .iloc expects integer index positions, not True/False.



```python
# breaks!
amazonbooks.iloc[ amazonbooks.isnull().sum(axis=1)>1 ,:]
```



---



📌 Step 1: `amazonbooks.isnull()`

This returns a DataFrame of the same shape as amazonbooks, but with:

* `True` where a cell is missing (i.e., NaN)

This code is filtering the dataset to return only the rows with absolutely no missing values.

* `amazonbooks.isnull()`	Returns a DataFrame of booleans, where each cell is `True` if it's missing (`NaN`) and `False` otherwise.

* `.sum(axis=1)`	Adds up all the `True` values per row. Since `True = 1`, this gives the count of NaNs per row.

* `== 0`	Checks where the count of missing values is 0 – i.e., no missing data in that row.

* Final result: `amazonbooks[...]`	Filters and returns only the fully complete rows – the cleanest part of your data!

```python
amazonbooks[amazonbooks.isnull().sum(axis=1)==0]
```

Imagine checking attendance sheets for classes. This line is like saying:

1. Grouping by a Boolean Condition:

* `amazonbooks.NumPages > 335.857585` returns a Boolean Series.

* This splits the data into two groups based on that condition.

2. `.mean()` on groupby object:

* Calculates the average of all numeric columns for each group.

* Ignores non-numeric columns like `Author`, `Title`, etc.

```python
amazonbooks.groupby(amazonbooks.NumPages > 335.857585).mean()
```



---



🧠 What's New Here?

✅ `numeric_only=True`:

* This parameter explicitly tells pandas to only compute the mean of numeric columns, avoiding any warning or ambiguity.

* Useful in newer versions of pandas (>=1.5) where some non-numeric columns might sneak into `.mean()` by accident.

* Ensures clean and consistent output.



```python
amazonbooks.groupby(amazonbooks.NumPages > 335.857585).mean(numeric_only=True)
```

🎯 What’s the Goal?

We want to find out:

> Is there a difference in average Amazon price between shorter books and longer books?


🔹 `amazonbooks[['Amazon Price']]`:

* This selects just the Amazon Price column as a DataFrame (not a Series).

* Why double brackets? Because we're passing a list of columns — even though it has only one item, it keeps it as a DataFrame (which is expected by `.groupby()` + `.mean()`).

🔹 .groupby(amazonbooks.NumPages > 335.857585):

* This creates two groups:

   * `False`: books with 335.857585 pages or fewer

   * `True`: books with more than 335.857585 pages

* We're essentially splitting books into “short” and “long” categories.

🔹 `.mean()`:

* For each group, it calculates the average Amazon Price.

* It tells us:

 > “Do longer books tend to cost more?”

```python
amazonbooks[['Amazon Price']].groupby(amazonbooks.NumPages > 335.857585).mean()
```



---



We want to create a new column in the dataset that categorizes each book as having above average page count or not.


🔹 `amazonbooks.NumPages > 335.857585`

* This is a boolean condition.

* It returns a `True` or `False` for each row, depending on whether that book has more pages than the average number of pages (in this case, ≈ 335.86).

* Example:

 * 500 pages → `True`

 * 200 pages → `False`

🔹 `amazonbooks['AboveAverageNumPages'] = ...`

* This creates a new column in the `amazonbooks` DataFrame.

* The column will store `True/False` values for each book.

* It becomes a flag or categorical label that we can use for grouping, filtering, and visualization.

```python
# add a new column!
amazonbooks['AboveAverageNumPages'] = amazonbooks.NumPages > 335.857585
```



---



🧠 What Are We Doing?
We are using grouping and aggregation to compare the average values (like price, number of pages, etc.) between two groups:

* Books with above-average number of pages

* Books with average or below-average number of pages

```python
amazonbooks.groupby('AboveAverageNumPages').mean(numeric_only=True)
```

🔸 `amazonbooks.groupby('AboveAverageNumPages')`

* This groups the dataset into two categories:

 * `False` → Books with 335 or fewer pages

 * `True` → Books with more than 335 pages

🔸 `.mean(numeric_only=True)`

* For each group (`True` and `False`), this computes the average (mean) of all numeric columns (like `Amazon Price`, `NumPages`, `Pub year`, etc.)

* `numeric_only=True` ensures that only numeric data is included in the mean calculation (e.g., ignores text like book titles or authors).

👉 To calculate the average cost per page for different types of books — hardcover vs paperback — and compare them.

```python
#amazonbooks['PricePerPage'] = amazonbooks.NumPages/amazonbooks['Amazon Price'] # oops! upside down
amazonbooks['PricePerPage'] = amazonbooks['Amazon Price']/amazonbooks.NumPages
amazonbooks[['PricePerPage','Hard_or_Paper']].groupby('Hard_or_Paper').mean()
```



---



🧠 Quick Refresher: What is Standard Deviation?

* Standard deviation tells us how much variation there is from the average (mean).

* A low standard deviation means the values are close to the mean (less spread out).

* A high standard deviation means the values are more spread out (more variability).

```python
# The "standard deviation" `.std()` function measure how spread around the data mean the data is
amazonbooks[['PricePerPage','Hard_or_Paper']].groupby('Hard_or_Paper').std()
```

✅ Interpretation:

* The standard deviation is slightly higher for paperback books.

* This means paperback books have a more varied (less consistent) cost per page.

* In contrast, hardcover books have more consistent pricing per page.
