```python
def super_vowels(text):
    VOWELS = "aeiou"
    result = ""
    print("Check vowels in string:")

    for ch in text:
        if ch.lower() in VOWELS:    
            result += ch.upper()
        elif ch.isalpha():        
            result += ch.lower()
        else:                         
            result += ch

    print("Converted vowels to UPPER and consonants to lower.")
    return result


print(super_vowels("qasim raza!"))
```

```python
def get_playing_board(size):
    board = ""
    for row in range(size):
      
        if row % 2 == 0:
            board += " " + "* " * (size // 2) + "\n"
        else:
            board += "* " * (size // 2) + "*\n"
    return board


board_of_5 = (
    " * * \n"
    "* * *\n"
    " * * \n"
    "* * *\n"
    " * * \n"
)

board_of_10 = (
    " * * * * *\n"
    "* * * * * \n"
    " * * * * *\n"
    "* * * * * \n"
    " * * * * *\n"
    "* * * * * \n"
    " * * * * *\n"
    "* * * * * \n"
    " * * * * *\n"
    "* * * * * \n"
)
assert get_playing_board(5) == board_of_5


print(get_playing_board(5))
```
