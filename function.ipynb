```python
data =[10,20,30,40,50]
def calculatemean(data):
    return sum(data)/len(data)

mean=calculatemean(data)
mean
```

```python
def calculate_std(data):
    mean=calculatemean(data)
    square_diff=[(x-mean)**2 for x in data]
    # print( square_diff)
    return(sum (square_diff)/(len(data)-1))**0.5
std=calculate_std(data)
std
```
